   __1-Visualisation des données__ :On commence par visualiser la distribution des valeurs de la colonne "Age" à l'aide d'histogrammes ou de tout autre graphique approprié. Cela nous aidera à repérer les éventuelles valeurs aberrantes.

__2-Définition des seuils__ : Déterminez les seuils supérieur et inférieur pour identifier les valeurs aberrantes. Vous pouvez utiliser des méthodes statistiques, telles que le calcul des quartiles ou l'écart interquartile (IQR), pour définir ces seuils. Par exemple, les valeurs inférieures à Q1 - 1,5 * IQR ou supérieures à Q3 + 1,5 * IQR peuvent être considérées comme des valeurs aberrantes.

```python
import numpy as np

# Calcul des quartiles pour les données d'entraînement
Q1_train = np.percentile(df_train['Age'].dropna(), 25)
Q3_train = np.percentile(df_train['Age'].dropna(), 75)

# Calcul des quartiles pour les données de test
Q1_test = np.percentile(df_test['Age'].dropna(), 25)
Q3_test = np.percentile(df_test['Age'].dropna(), 75)

# Calcul de l'écart interquartile (IQR) pour les données d'entraînement
IQR_train = Q3_train - Q1_train

# Calcul de l'écart interquartile (IQR) pour les données de test
IQR_test = Q3_test - Q1_test

# Définition des seuils supérieur et inférieur pour les données d'entraînement
seuil_inferieur_train = Q1_train - 1.5 * IQR_train
seuil_superieur_train = Q3_train + 1.5 * IQR_train

# Définition des seuils supérieur et inférieur pour les données de test
seuil_inferieur_test = Q1_test - 1.5 * IQR_test
seuil_superieur_test = Q3_test + 1.5 * IQR_test

# Affichage des seuils pour les données d'entraînement
print("Seuils pour les données d'entraînement:")
print("Seuil inférieur:", seuil_inferieur_train)
print("Seuil supérieur:", seuil_superieur_train)

# Affichage des seuils pour les données de test
print("Seuils pour les données de test:")
print("Seuil inférieur:", seuil_inferieur_test)
print("Seuil supérieur:", seuil_superieur_test)
```

__3-Identification des valeurs aberrantes__  :On a identifié les lignes du dataset qui contiennent des valeurs aberrantes en comparant chaque valeur de la colonne "Age" aux seuils définis précédemment. On peut utiliser des opérations de filtrage ou de masquage pour extraire ces lignes.

```python
# Filtrage des valeurs aberrantes dans les données d'entraînement
valeurs_aberrantes_train = df_train[(df_train['Age'] < seuil_inferieur_train) | (df_train['Age'] > seuil_superieur_train)]

# Filtrage des valeurs aberrantes dans les données de test
valeurs_aberrantes_test = df_test[(df_test['Age'] < seuil_inferieur_test) | (df_test['Age'] > seuil_superieur_test)]
```

```python
import matplotlib.pyplot as plt

# Visualisation des valeurs aberrantes identifiées pour les données d'entraînement
plt.scatter(df_train.index, df_train['Age'], color='b', label='Non aberrantes (train)')
plt.scatter(valeurs_aberrantes_train.index, valeurs_aberrantes_train['Age'], color='r', label='Aberrantes (train)')

# Visualisation des valeurs aberrantes identifiées pour les données de test
plt.scatter(df_test.index, df_test['Age'], color='g', label='Non aberrantes (test)')
plt.scatter(valeurs_aberrantes_test.index, valeurs_aberrantes_test['Age'], color='m', label='Aberrantes (test)')

plt.xlabel('Index')
plt.ylabel('Age')
plt.title('Valeurs aberrantes de la colonne "Age"')
plt.legend()
plt.show()
```

__4-Gestion des valeurs aberrantes__ Une fois les valeurs aberrantes identifiées, vous pouvez choisir de les traiter de différentes manières. Voici quelques options :__Suppression__ , __Remplacement__ , __Imputation__

__Imputation__ : Une autre approche consiste à imputer les valeurs manquantes en utilisant des méthodes d'imputation plus avancées, telles que la régression, qui prennent en compte les autres caractéristiques du dataset pour estimer les valeurs manquantes de manière plus précise.

# 5- L'ingénierie de fonctionnalités

L'ingénierie de fonctionnalités consiste à **transformer les données brutes** (telles que l'âge, le sexe, la classe de billet, etc.) **en fonctionnalités qui peuvent être utilisées pour former un modèle de prédiction**. Les fonctionnalités peuvent inclure des informations supplémentaires dérivées des données d'origine ou des transformations des données existantes pour mieux représenter les informations contenues dans la dataset.

Par exemple, pour la dataset Titanic, on peut créer des fonctionnalités telles que le titre du passager (M. ou Mme) en utilisant le nom, la taille de la famille en combinant le nombre de frères et sœurs/époux et de parents/enfants, ou encore l'emplacement de la cabine en utilisant le numéro de cabine.

L'ingénierie de fonctionnalités est une **étape importante dans la construction de modèles de prédiction précis et robustes**, car elle permet d'extraire des informations utiles à partir des données brutes et de fournir une représentation plus complète et pertinente de la dataset.

## 1) Techniques d'extraction de fonctionnalités

###  __1) Extraction du titre à partir de la variable nominale "Nom".

Il est observé que les noms des passagers contiennent des titres tels que "Mrs.", "Mr.", etc. Il serait pertinent de extraire cette information et de créer une nouvelle caractéristique (feature) à partir de ces titres.

Dans le code suivant, nous extrayons la fonction Titre à l'aide d'expressions régulières. Le modèle RegEx (\w+\.) correspond au premier mot qui se termine par un point dans la fonction Nom. L'indicateur expand=False renvoie un DataFrame.

```python
# Extraction du titre à partir du nom pour chaque DataFrame
df_train['Title'] = df_train.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
df_test['Title'] = df_test.Name.str.extract(' ([A-Za-z]+)\.', expand=False)

# Tableau croisé des titres et du sexe dans le DataFrame d'entraînement
pd.crosstab(df_train['Title'], df_train['Sex'])
```

Nous pouvons remplacer de nombreux titres par un nom plus courant ou les classer comme rares.

```python
for dataset in [df_train, df_test]:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mrs', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')

df_train[['Title', 'Survived']].groupby(['Title'], as_index=False).mean()
```

### __2) Discrétisation de la variable continue "Âge".__

Nous allons créer là aussi plusieurs catégories d’age selon la variable Age:

Les bébés : de 0 a 3 ans                      
Les enfants: de 3 à 15 ans  
Les adultes de 15 à 60 ans  
Les « vieux » de plus de 60 ans  

```python
import pandas as pd
# Remplacement des valeurs manquantes dans la colonne 'Age' par la moyenne
age_train = df_train['Age'].fillna(df_train['Age'].mean())
age_test = df_test['Age'].fillna(df_test['Age'].mean())

# Création de la colonne 'catAge' en fonction des catégories d'âge pour train_df
catAge_train = []
for i in range(df_train.shape[0]):
    if age_train[i] <= 3:
        catAge_train.append("bebe")
    elif age_train[i] > 3 and age_train[i] < 15:
        catAge_train.append("enfant")
    elif age_train[i] >= 15 and age_train[i] <= 60:
        catAge_train.append("adulte")
    else:
        catAge_train.append("vieux")

# Ajout de la colonne 'catAge' au DataFrame train_df
df_train['catAge'] = catAge_train

# Création de la colonne 'catAge' en fonction des catégories d'âge pour test_df
catAge_test = []
for i in range(df_test.shape[0]):
    if age_test[i] <= 3:
        catAge_test.append("bebe")
    elif age_test[i] > 3 and age_test[i] < 15:
        catAge_test.append("enfant")
    elif age_test[i] >= 15 and age_test[i] <= 60:
        catAge_test.append("adulte")
    else:
        catAge_test.append("vieux")

# Ajout de la colonne 'catAge' au DataFrame test_df
df_test['catAge'] = catAge_test

# Affichage des premières lignes du DataFrame train_df avec la colonne 'catAge'
df_train.head(25)
```

# Techniques de Feature Selection


Il existe plusieurs techniques de sélection de fonctionnalités que vous pouvez utiliser pour obtenir de meilleures prédictions. Voici une technique couramment utilisée : la sélection par importance des variables  
Pour sélectionner les meilleures fonctionnalités et évaluer les performances du modèle, vous pouvez utiliser la méthode Recursive Feature Elimination (RFE) .

__Ce code entraîne un modèle de classification RandomForest sur un ensemble de formation, calcule l'importance des variables, sélectionne les caractéristiques les plus importantes, et crée de nouveaux ensembles de données avec ces caractéristiques pour l'ensemble d'entraînement et l'ensemble de test.__

```python
from sklearn.ensemble import RandomForestClassifier

# Création d'un modèle de classification RandomForest
model = RandomForestClassifier()

# Entraînement du modèle sur l'ensemble d'entraînement
model.fit(X_train, y_train)

# Obtention de l'importance des variables
importances = model.feature_importances_

# Création d'un DataFrame pour visualiser les importances des variables
feature_importances = pd.DataFrame({'Feature': X_train.columns, 'Importance': importances})

# Tri des variables par importance décroissante
feature_importances = feature_importances.sort_values('Importance', ascending=False)

# Sélection des meilleures variables (par exemple, les 5 premières)
selected_features = feature_importances.head(5)['Feature'].tolist()

# Sélection des caractéristiques et de la variable cible pour l'ensemble de formation et l'ensemble de test
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]

# Visualisation des importances des variables
print(feature_importances)

# Affichage de l'ensemble de données sélectionné pour l'ensemble d'entraînement
print("Ensemble de données sélectionné pour l'ensemble d'entraînement :")
print(X_train_selected.head())

# Affichage de l'ensemble de données sélectionné pour l'ensemble de test
print("Ensemble de données sélectionné pour l'ensemble de test :")
print(X_test_selected.head())
```

Le codes fourni est destiné à effectuer la sélection des fonctionnalités les plus pertinentes pour le modèle. Pour évaluer si la sélection de fonctionnalités a conduit à de meilleures prédictions, on va entraîner un modèle de prédiction sur les données sélectionnées et évaluer ses performances.

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

# Création d'un modèle de classification RandomForest
model = RandomForestClassifier()

# Entraînement du modèle sur les données sélectionnées
model.fit(X_train_selected, y_train)

# Prédiction sur l'ensemble de test
y_pred = model.predict(X_test_selected)

# Évaluation des performances
accuracy = accuracy_score(y_test, y_pred)
print("Précision du modèle : {:.2f}%".format(accuracy * 100))

# Validation croisée pour une évaluation plus robuste
cv_scores = cross_val_score(model, X_train_selected, y_train, cv=5)
print("Précision moyenne avec validation croisée : {:.2f}%".format(cv_scores.mean() * 100))
```

Dans cet exemple, nous utilisons la classe RFECV pour effectuer une sélection de fonctionnalités en utilisant la méthode Recursive Feature Elimination. Le modèle RandomForestClassifier est utilisé comme estimateur de base. La méthode fit_transform est utilisée pour sélectionner les caractéristiques les plus pertinentes sur l'ensemble d'entraînement, puis le modèle est entraîné sur ces fonctionnalités sélectionnées. Ensuite, les mêmes fonctionnalités sélectionnées sont utilisées pour prédire les valeurs sur l'ensemble de test. Les performances du modèle sont évaluées à l'aide de la métrique d'exactitude (accuracy) et une validation croisée est effectuée pour obtenir une estimation plus robuste des performances.

```python
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier

# Création d'un modèle de classification RandomForest
model = RandomForestClassifier()

# Sélection des caractéristiques en utilisant la méthode Recursive Feature Elimination (RFE)
rfe = RFECV(estimator=model, cv=5)
X_train_selected = rfe.fit_transform(X_train, y_train)

# Entraînement du modèle sur les données sélectionnées
model.fit(X_train_selected, y_train)

# Prédiction sur l'ensemble de test en utilisant les mêmes fonctionnalités sélectionnées
X_test_selected = rfe.transform(X_test)
y_pred = model.predict(X_test_selected)

# Évaluation des performances
accuracy = accuracy_score(y_test, y_pred)
print("Précision du modèle : {:.2f}%".format(accuracy * 100))
```

# Pourquoi Feature Selection est importante ? 

La sélection de fonctionnalités est importante pour plusieurs raisons :

Réduction de la dimensionnalité : Lorsque vous avez un grand nombre de fonctionnalités, certaines d'entre elles peuvent être redondantes ou peu informatives. La sélection de fonctionnalités permet de réduire la dimensionnalité des données en identifiant les caractéristiques les plus pertinentes .

Amélioration de la précision : En éliminant les caractéristiques bruitées ou peu importantes, la sélection de fonctionnalités peut aider à améliorer la précision du modèle. 

Interprétabilité : En sélectionnant les caractéristiques les plus importantes, il est plus facile d'interpréter et de comprendre les résultats du modèle. 

Réduction du surapprentissage : Lorsque le nombre de fonctionnalités est élevé par rapport à la taille de l'ensemble de données, il existe un risque de surapprentissage, c'est-à-dire que le modèle peut apprendre des relations insignifiantes ou bruitées. La sélection de fonctionnalités permet de réduire ce risque en se concentrant sur les caractéristiques les plus discriminantes et en évitant la surcharge d'informations inutiles.

# 7- Application des algorithmes de machine learning à la dataset "titanic 


```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Initialisation des modèles
svm = SVC()
knn = KNeighborsClassifier(n_neighbors=5)
gnb = GaussianNB()
lr = LogisticRegression()
dt = DecisionTreeClassifier()
rfc = RandomForestClassifier()
sgd = SGDClassifier()
gbc = GradientBoostingClassifier()

models = [svm, knn, gnb, lr, dt, rfc, sgd, gbc]
names = ['SVM', 'KNN', 'Gaussian Naive Bayes', 'Logistic Regression', 'Decision Tree Classifier', 'Random Forest Classifier', 'Stochastic Gradient Descent', 'Gradient Boosting Classifier']

accuracies = []

# Validation croisée des modèles et calcul de leur précision moyenne
for name, model in zip(names, models):
    cv_score = cross_val_score(model, X_train, y_train, cv=5)
    print("{} - Exactitude moyenne : {:.2f} %".format(name, np.mean(cv_score)*100))

    # Calcul de l'exactitude du modèle sur les données de test
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)
    print("{} - Exactitude sur les données de test : {:.2f} %".format(name, accuracy*100))

    # Calcul de la matrice de confusion et du rapport de classification
    cm = confusion_matrix(y_test, y_pred)
    print("{} - Matrice de confusion : \n{}".format(name, cm))
    cr = classification_report(y_test, y_pred)
    print("{} - Rapport de classification : \n{}".format(name, cr))
```

```python
# Trouver le modèle le plus performant
best_index = np.argmax(accuracies)
best_model = models[best_index]
best_name = names[best_index]
print("Le meilleur classifieur est {} avec une exactitude de {:.2f} % sur les données de test.".format(best_name, accuracies[best_index]*100))
```
